```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car data.csv')
df.head()
```

```python
for feature in df.columns:
    print("{} : {}".format(feature , df[feature].nunique()))
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

```python
final_df = df.drop(columns=['Car_Name'])
```

```python
final_df.columns
```

```python
final_df['Current_Year'] = 2020
final_df.head()
```

```python
final_df['no_year'] = final_df['Current_Year'] - final_df['Year']
```

```python
final_df.head()
```

```python
final_df.drop(columns=['Year','Current_Year'],inplace=True)
```

```python
final_df.head()
```

```python
final_df = pd.get_dummies(final_df, drop_first=True)
```

```python
final_df.head()
```

```python
final_df.corr()
```

```python
sns.pairplot(final_df)
```

```python
%matplotlib inline
```

```python
corrmat = final_df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(13,13))
#plot heatmap
sns.heatmap(final_df[top_corr_features].corr() , annot=True , cmap='RdYlGn') #uses pearson's correaltion 
plt.show();
```

```python
final_df.head(2)
```

```python
#independent and dependent features
X = final_df.iloc[:,1:]
y = final_df.iloc [:,0]                                                                       
```

```python
X.head()
```

```python
#feature importance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
plt.style.use('fivethirtyeight')
```

```python
#plot graph for feature importance
feat_importance = pd.Series(model.feature_importances_ , index=X.columns)
feat_importance.nlargest(5).plot(kind='barh')
plt.xlabel('Score')
plt.ylabel('Feature')
plt.title('Top 5 features')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print("X_train ; {}   X_test : {}\ny_train : {}     y_test : {}".format(X_train.shape, X_test.shape, y_train.shape ,y_test.shape))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
##hyper-parmeters
n_estimators = [int(x) for x in np.linspace(start=100 , stop=1200 ,num=12)]
max_features = ['auto' , 'sqrt']
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
min_sample_split = [2,5,10,15,100]
min_sample_leaf = [1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#create the random grid,
random_grid = {'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_sample_split,
               'min_samples_leaf':min_sample_leaf               
}
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf , param_distributions=random_grid , n_jobs=1 ,scoring='neg_mean_squared_error' ,n_iter=10 , cv=5 ,verbose=2 ,random_state=42)
```

```python
rf_random.fit(X_train,y_train)
```

```python
y_pred = rf_random.predict(X_test)
```

```python
y_pred
```

```python
sns.distplot(y_test-y_pred)
```

```python
plt.scatter(y_test , y_pred)
```
